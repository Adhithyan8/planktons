```python
import os
import matplotlib.pyplot as plt
```

```python
year = 2013
```

```python
# how many folders are in data/2014?
print(len(os.listdir(f'data/{year}')))
```

```python
categories = {}
for category in os.listdir(f'data/{year}'):
    categories[category] = len(os.listdir(f'data/{year}/'+category))
```

```python
# plot the number of images in each category
plt.figure(figsize=(20, 5))
plt.bar(range(len(categories)), list(categories.values()), align='center')
plt.xticks(range(len(categories)), list(categories.keys()), rotation='vertical')
# log scale
plt.yscale('log')
plt.show()
```
