```python
from datasets import load_dataset
import random
import matplotlib.pyplot as plt
```

```python
train_dataset = load_dataset("planktons_dataset", "2013-14", split="train")
test_dataset = load_dataset("planktons_dataset", "2013-14", split="test")
```

```python
# choose 5 random classes and plot 5 random images from each of them
num_classes = 5
num_samples = 5

classes = random.sample(train_dataset.features["label"].names, num_classes)
class_ids = [train_dataset.features["label"].names.index(c) for c in classes]
```

```python
class_ids
```

```python
fig, axs = plt.subplots(num_classes, num_samples, figsize=(10, 10))
for i, cls_id in enumerate(class_ids):
    samples = random.sample(train_dataset.filter(lambda x: x["label"] == cls_id)["image"], num_samples)
    for j, sample in enumerate(samples):
        axs[i, j].imshow(sample)
        if i == j:
            axs[i, j].set_title(classes[i], fontsize=10)
        axs[i, j].axis("off")

plt.tight_layout()
plt.savefig("sample_train.png")
```
